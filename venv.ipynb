```python
import requests
from bs4 import BeautifulSoup

# Step 1: Set the target URL
url = "https://www.healthforcalifornia.com/covered-california/insurance"

# Step 2: Send a GET request
response = requests.get(url)

# Step 3: Check if the request was successful
if response.status_code == 200:
    # Step 4: Parse the page content
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Step 5: Find articles (adjust selectors based on CMS.gov structure)
    articles = soup.find_all('div', class_='views-row')
    
    # Step 6: Extract and display data
    for article in articles:
        title = article.find('a').get_text(strip=True)
        summary = article.find('div', class_='field-content').get_text(strip=True) if article.find('div', class_='field-content') else "No summary available"
        
        print(f"Title: {title}")
        print(f"Summary: {summary}")
        print("-" * 50)
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```
